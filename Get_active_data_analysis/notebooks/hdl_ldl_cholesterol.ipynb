```python
data.head()
def age_group(x):
    if x['AGE'] <=25  :
        return 'G1'
    elif (x['AGE']<=30)&(x['AGE']>25)  :
        return 'G2'
    elif (x['AGE']<=35)&(x['AGE']>30)  :
        return 'G3'
    elif (x['AGE']<=40)&(x['AGE']>35)  :
        return 'G4'
    elif (x['AGE']<=45)&(x['AGE']>40)  :
        return 'G5'
    elif (x['AGE']<=50)&(x['AGE']>45)  :
        return 'G6'
    else :
        return 'old'
    
data['age_group'] = data.apply(age_group,axis=1)
```
